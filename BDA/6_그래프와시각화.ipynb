### **중요!** figure 생성 및 axes 객체  생성 방법-2 (b)

```python
# fig, axes 동시 생성
fig, axes = plt.subplots(2, 2)

axes[0,0].hist(np.random.standard_normal(100), bins=20, color="black", alpha=0.3);
axes[0,1].scatter(np.arange(30), np.arange(30) + 3 * np.random.standard_normal(30));
axes[1,0].plot(np.random.standard_normal(50).cumsum(), color="black",linestyle="dashed")
```

## 2. 다양한 subplots 옵션-sharex,sharey,wspace,hspace

![](images/subplot옵션.png)

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)  # x축 눈금, y축 눈금 공유
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.standard_normal(500), bins=50, color="black", alpha=0.5)  #color, alpha 
# 그래프간 가로 간격, 세로 간격
fig.subplots_adjust(wspace=0.1, hspace=0.03)    # adjust wspace, hspace
```

## 6. 레스토랑 Tip 예제

```python
tips = pd.read_csv("examples/tips.csv")
tips
```

```python
tips.shape
```

```python
# crosstab 중요 X. 도수 확인
party_counts = pd.crosstab(tips["day"], tips["size"]) # 두 columns에 대해서 도수를 반환
party_counts
```

```python
party_counts = party_counts.reindex(index=["Thur", "Fri", "Sat", "Sun"])
party_counts
```

```python
# loc이므로 2:5가 인덱스순서(정수)가 아니라 인덱스 라벨을 의미
party_counts = party_counts.loc[:, 2:5]  # 2,3,4,5만 많음(인덱스는 슬라이싱 시에 end point 포함)
```

```python
party_counts
```

```python
# Normalize to sum to 1,정규화
party_pcts = party_counts.div(party_counts.sum(axis="columns"), axis="index")  # 요소 값 / 각 행의 총합
party_pcts
```

```python
party_pcts.plot.bar(stacked=True)
```

```python
macro = pd.read_csv("examples/macrodata.csv")
data = macro[["cpi", "m1", "tbilrate", "unemp"]]
trans_data = np.log(data).diff().dropna()  #diff: 행 간 차이값
trans_data.tail()
```

```python
# regplot-산포도,회귀선
ax = sns.regplot(x="m1", y="unemp", data=trans_data)
ax.set_title("Changes in log(m1) versus log(unemp)")
```

```python
# pairplot: olumn 들의 산포도를 그리고 대각 성분은 히스토그램을 그림
# diag_kind=‘kde’ (히스토그램 대신 커널 추정 밀도함수)
sns.pairplot(trans_data, diag_kind="kde", plot_kws={"alpha": 0.2})   #plot_kws: plot 할 때 option과 arguments (keyward arguments)
```

## 8. **중요!** 시본으로 subplot 그리기

### subplots+ seaborn  (seaborn으로 subplots 그리기,  ax option)

```python
fig,ax=plt.subplots(1,2)
sns.boxplot(x="tip_pct", y="day", data=tips[tips.tip_pct < 0.5],ax=ax[0])
sns.barplot(x="day", y="tip_pct", hue='smoker', data=tips[tips.tip_pct < 1])
```

### Catplot (categorical plot),   row와 col로 category 결정


```python
sns.catplot(x="day", y="tip_pct", hue="time", col="smoker", kind="bar", data=tips[tips.tip_pct < 1])
# smoker열의 값에 따라 따라 서로 다른 서브플롯 생성
```

```python
sns.catplot(x="day", y="tip_pct", row="time",col="smoker",kind="bar", data=tips[tips.tip_pct < 1])
```

```python
sns.catplot(x="tip_pct", y="day", kind="box",data=tips[tips.tip_pct < 0.5])  
```

```python
sns.boxplot(x="tip_pct", y="day", data=tips[tips.tip_pct < 0.5])  #FacetGrid vs AxesSubplot
```

### 패싯그리드 (FacetGrid): 시본 객체로, multiplot을 쉽게 그릴 수 있게 함
다만 map을 쓰는것보단 보통 catplot을 직접 활용하는 편이 더 선호됨

```python
facet=sns.FacetGrid(tips,row="time", col="smoker" ) # data=tips
facet.map(sns.histplot,'total_bill')    # total_bill에 대한 히스토그램
```

```python
facet=sns.FacetGrid(data=tips[tips.tip_pct < 1],row="time", col="smoker" )
facet.map(sns.barplot, "day", "tip_pct")
```
